# Extracted Data

```python
import pandas as pd

# Path to the source Excel file (extracted data)
source_file_path = r'C:\Users\Shruti Khivansara\Downloads\consumer data.xlsx'

# Path to the mapping Excel file
mapping_file_path = r'C:\Users\Shruti Khivansara\Downloads\smart360.xlsx'

# Read the source and mapping Excel files into pandas DataFrames
try:
    source_df = pd.read_excel(source_file_path)
    mapping_df = pd.read_excel(mapping_file_path)
    
    # Perform the mapping based on common columns
    merged_df = pd.merge(source_df, mapping_df, how='inner', on='Consumer ID')
    
    # Display the merged DataFrame
    print(merged_df)
    
    # Now you can further process the merged data as needed
except FileNotFoundError:
    print(f"File not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
